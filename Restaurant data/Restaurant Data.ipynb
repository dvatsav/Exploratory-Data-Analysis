```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%config InlineBackend.figure_format = 'svg'
from pylab import rcParams
rcParams['figure.figsize'] = 5, 4
```

```python
df = pd.read_csv("zomato.csv", encoding = "latin1")
```

```python
df['Has Table booking'] = df['Has Table booking'].map({'Yes' : 1, 'No' : 0})
df['Has Online delivery'] = df['Has Online delivery'].map({'Yes' : 1, 'No' : 0})
df['Is delivering now'] = df['Is delivering now'].map({'Yes' : 1, 'No' : 0})
df['Switch to order menu'] = df['Switch to order menu'].map({'Yes' : 1, 'No' : 0})

del df['Longitude']
del df['Latitude']
del df['Address']
del df['Locality']
del df['Locality Verbose']

city_map = {}
i = 1
for city in df['City'].unique():
    city_map[city] = i
    i += 1
df['City'] = df['City'].map(city_map)

currency_map = {}
i = 1
for currency in df['Currency'].unique():
    currency_map[currency] = i
    i += 1
df['Currency'] = df['Currency'].map(currency_map)
```

```python
features = ['Votes', 'Has Table booking', 'Has Online delivery', 'Is delivering now', 'Price range', 'Aggregate rating']
corr_matrix = df[features].corr()
sns.heatmap(corr_matrix)
```
